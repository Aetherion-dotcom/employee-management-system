```python
import pandas as pd
import glob
```

```python
enrolment_files = glob.glob('api_data_aadhar_enrolment/api_data_aadhar_enrolment_*.csv')
demographic_files = glob.glob('api_data_aadhar_demographic/api_data_aadhar_demographic_*.csv')

enrolment_dfs = []
for file in enrolment_files:
    enrolment_dfs.append(pd.read_csv(file))
df_enrolment = pd.concat(enrolment_dfs, ignore_index=True)

demographic_dfs = []
for file in demographic_files:
    demographic_dfs.append(pd.read_csv(file))
df_demographic = pd.concat(demographic_dfs, ignore_index=True)

print("df_enrolment Head:")
print(df_enrolment.head())
print("\ndf_enrolment Info:")
df_enrolment.info()

print("\n\ndf_demographic Head:")
print(df_demographic.head())
print("\ndf_demographic Info:")
df_demographic.info()
```

```python
df_merged = pd.merge(df_enrolment, df_demographic, on=['date', 'state', 'district', 'pincode'], how='inner')

print("df_merged Head:")
print(df_merged.head())

print("\ndf_merged Info:")
df_merged.info()
```

```python
df_merged['date'] = pd.to_datetime(df_merged['date'], format='%d-%m-%Y')

print("df_merged Info after date conversion:")
df_merged.info()

print("\nMissing values in df_merged:")
print(df_merged.isnull().sum())

print("\nDescriptive statistics for df_merged:")
print(df_merged.describe())
```

```python
df_daily_enrolment_trends = df_merged.groupby('date')[['age_0_5', 'age_5_17', 'age_18_greater']].sum().reset_index()
df_daily_demographic_trends = df_merged.groupby('date')[['demo_age_5_17', 'demo_age_17_']].sum().reset_index()

print("\nDaily Enrolment Trends Head:")
print(df_daily_enrolment_trends.head())

print("\nDaily Demographic Trends Head:")
print(df_daily_demographic_trends.head())

print("\nDescriptive statistics for Daily Enrolment Trends:")
print(df_daily_enrolment_trends.describe())

print("\nDescriptive statistics for Daily Demographic Trends:")
print(df_daily_demographic_trends.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(15, 7))
plt.plot(df_daily_enrolment_trends['date'], df_daily_enrolment_trends['age_0_5'], label='Age 0-5', marker='.')
plt.plot(df_daily_enrolment_trends['date'], df_daily_enrolment_trends['age_5_17'], label='Age 5-17', marker='.')
plt.plot(df_daily_enrolment_trends['date'], df_daily_enrolment_trends['age_18_greater'], label='Age 18+', marker='.')
plt.title('Daily Aadhaar Enrolment Trends by Age Group')
plt.xlabel('Date')
plt.ylabel('Number of Enrolments')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot daily demographic update trends
plt.figure(figsize=(15, 7))
plt.plot(df_daily_demographic_trends['date'], df_daily_demographic_trends['demo_age_5_17'], label='Demographic Update Age 5-17', marker='.')
plt.plot(df_daily_demographic_trends['date'], df_daily_demographic_trends['demo_age_17_'], label='Demographic Update Age 17+', marker='.')
plt.title('Daily Aadhaar Demographic Update Trends by Age Group')
plt.xlabel('Date')
plt.ylabel('Number of Demographic Updates')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def find_outliers_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    return outliers

# 1. Calculate 'Total Enrolments' for df_daily_enrolment_trends
df_daily_enrolment_trends['total_enrolments'] = df_daily_enrolment_trends['age_0_5'] + df_daily_enrolment_trends['age_5_17'] + df_daily_enrolment_trends['age_18_greater']

# 2. Calculate 'Total Demographic Updates' for df_daily_demographic_trends
df_daily_demographic_trends['total_demographic_updates'] = df_daily_demographic_trends['demo_age_5_17'] + df_daily_demographic_trends['demo_age_17_']

# 3. Calculate 'Total Enrolments' for df_enrolment_by_state
df_enrolment_by_state['total_enrolments'] = df_enrolment_by_state['age_0_5'] + df_enrolment_by_state['age_5_17'] + df_enrolment_by_state['age_18_greater']

# 4. Calculate 'Total Demographic Updates' for df_demographic_by_state
df_demographic_by_state['total_demographic_updates'] = df_demographic_by_state['demo_age_5_17'] + df_demographic_by_state['demo_age_17_']

# 5. Calculate 'Total Enrolments' for df_enrolment_by_district
df_enrolment_by_district['total_enrolments'] = df_enrolment_by_district['age_0_5'] + df_enrolment_by_district['age_5_17'] + df_enrolment_by_district['age_18_greater']

# 6. Calculate 'Total Demographic Updates' for df_demographic_by_district
df_demographic_by_district['total_demographic_updates'] = df_demographic_by_district['demo_age_5_17'] + df_demographic_by_district['demo_age_17_']

# 7 & 8. Identify and print outliers for each total column
print("\nOutliers in Daily Enrolment Trends (total_enrolments):")
print(find_outliers_iqr(df_daily_enrolment_trends, 'total_enrolments'))

print("\nOutliers in Daily Demographic Update Trends (total_demographic_updates):")
print(find_outliers_iqr(df_daily_demographic_trends, 'total_demographic_updates'))

print("\nOutliers in Enrolment by State (total_enrolments):")
print(find_outliers_iqr(df_enrolment_by_state, 'total_enrolments'))

print("\nOutliers in Demographic Update by State (total_demographic_updates):")
print(find_outliers_iqr(df_demographic_by_state, 'total_demographic_updates'))

print("\nOutliers in Enrolment by District (total_enrolments):")
print(find_outliers_iqr(df_enrolment_by_district, 'total_enrolments'))

print("\nOutliers in Demographic Update by District (total_demographic_updates):")
print(find_outliers_iqr(df_demographic_by_district, 'total_demographic_updates'))
```

## Analyze State-wise Aadhaar Coverage

### Subtask:
Calculate the total Aadhaar enrolments for each state using the existing df_enrolment_by_state DataFrame. Then, identify and visualize the states with the highest and lowest total enrolments to understand the distribution of Aadhaar coverage across different regions.


**Reasoning**:
The subtask requires identifying and visualizing the top 10 and bottom 10 states by total Aadhaar enrolments. This involves sorting the `df_enrolment_by_state` DataFrame and then generating two bar plots to display these findings.



```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Identify top 10 states by total enrolments
top_10_enrolment_states = df_enrolment_by_state.sort_values(by='total_enrolments', ascending=False).head(10)

# Identify bottom 10 states by total enrolments
bottom_10_enrolment_states = df_enrolment_by_state.sort_values(by='total_enrolments', ascending=True).head(10)

print("\nTop 10 States by Total Aadhaar Enrolments:")
print(top_10_enrolment_states[['state', 'total_enrolments']])

print("\nBottom 10 States by Total Aadhaar Enrolments:")
print(bottom_10_enrolment_states[['state', 'total_enrolments']])

# Plot for Top 10 States
plt.figure(figsize=(12, 7))
sns.barplot(x='state', y='total_enrolments', data=top_10_enrolment_states, palette='viridis')
plt.title('Top 10 States by Total Aadhaar Enrolments')
plt.xlabel('State')
plt.ylabel('Total Enrolments')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot for Bottom 10 States
plt.figure(figsize=(12, 7))
sns.barplot(x='state', y='total_enrolments', data=bottom_10_enrolment_states, palette='magma')
plt.title('Bottom 10 States by Total Aadhaar Enrolments')
plt.xlabel('State')
plt.ylabel('Total Enrolments')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
